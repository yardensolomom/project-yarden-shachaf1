הקוד הבא נעשה על מנת להבין את כמות המשתתפים במחקר, נעזרנו בצ'אט GPT. 
</h1> 


```python


# Initialize participant counter
participant_counter = 1

# Initialize the participant_number column with zeros
dating_df['participant_number'] = 0
dating_df['income'] = dating_df['income'].fillna(0)
# Iterate through the rows starting from the first row
for i in range(1, len(dating_df)):
    # Check if the income value is the same as the previous row
  
    if dating_df.at[i,'income'] != dating_df.at[i-1,'income'] or dating_df.at[i, 'age'] != dating_df.at[i-1, 'age'] or dating_df.at[i, 'goal'] != dating_df.at[i-1, 'goal'] or dating_df.at[i, 'career'] != dating_df.at[i-1, 'career']:
        # Increment the participant counter
        participant_counter += 1
    # Assign the participant number
    dating_df.at[i, 'participant_number'] = participant_counter

# Set the participant number for the first row
dating_df.at[0, 'participant_number'] = 1

dating_df


```

```python
dating_df["genders"]=dating_df["gender"].replace({0:'female',1:"male"})
dating_df['genders'].value_counts()
plt.figure(1,figsize=(5,5))
plt.title("participents: females vs males",fontsize=20)
dating_df['genders'].value_counts().plot.pie(autopct='%1.1f%%',shadow = True, explode = (0.01, 0.1), colors = ['lightblue', 'pink'])
plt.ylabel("")
plt.show()
    
```

<h1 style="font-family: 'Arial', sans-serif; font-size: 15px; direction: rtl; text-align: right; font-weight: bold;"> 
על מנת לקבל תמונה על ההערכות שקיבל כל משתתף, סידרנו את הטבלה כך שלכל משתתף מופיע ממוצע הפידבקים שקיבל עבור כל אחד מהקריטריונים.
    נעזרנו בצ'אט GPT. 
</h1>  

```python



# Identify numeric columns and non-numeric columns to be included in the result
numeric_columns = ['age','gender', 'goal','income', 'dec', 'attr', 'sinc', 'intel', 'fun', 'amb', 'shar', 'like', 'prob', 'met']
non_numeric_columns = ['career']

# Convert non-numeric values in numeric columns to NaN
for col in numeric_columns:
    dating_df[col] = dating_df[col].replace(['NA', 'inf', '-inf'], np.nan)
    dating_df[col] = pd.to_numeric(dating_df[col], errors='coerce')

# Group by 'participant_number' and calculate the mean for numeric columns
grouped = dating_df.groupby('participant_number')[numeric_columns].mean().reset_index()

# Convert averages to integers
grouped = grouped.round(0).astype('Int64')


# Add non-numeric columns to the result by taking the first non-null value for each participant
for col in non_numeric_columns:
    grouped[col] = dating_df.groupby('participant_number')[col].apply(lambda x: x.dropna().iloc[0] if not x.dropna().empty else np.nan).reset_index(drop=True)

 #Create the final DataFrame with specified columns
result_df = grouped[['participant_number'] + non_numeric_columns+ numeric_columns]

dating_df=result_df
result_df
```

ציפינו שאנשים עם הכנסה גבוהה ידורגו גבוה יותר והופתענו לגלות שדווקא אנשים עם הכנסה ממוצעת דורגו יותר גבוה ברוב הקטגוריות המשמעות שהכנסה לב דווקא השפיעה על הבחירה.

```python
def categorize_income(income):
    if pd.isna(income):
        return 'Unknown' 
    elif income > 50000:
        return 'income>50k'
    elif 30000 <= income <= 49999: 
        return '30k<income<50k' 
    elif 10000 <= income <= 29999:
        return '10k<income<30k'
    elif 1 <= income <= 9999: 
        return 'income<10k' 
    elif income == 0:
        return 'Unemployed' 
    else: 
        return 'Unemployed' 
dating_df=dating_df.copy()
dating_df['income_category']=dating_df['income'].apply(categorize_income)
average_ratings_by_income = dating_df.groupby('income_category')[['attr', 'sinc', 'intel', 'fun', 'amb', 'shar', 'like', 'prob']].mean()
average_ratings_by_feature= average_ratings_by_income.T  
average_ratings_by_feature.plot(kind='bar', figsize=(24, 20), cmap='tab20b')
ax.set_ylim(0,10)
plt.title('Average Ratings by Feature',fontsize=25) 
plt.xlabel('Rating Feature',fontsize=22) 
plt.ylabel('Average Rating',fontsize=22) 
plt.xticks(rotation=45)
plt.tick_params(axis='x',labelsize=20)
plt.tick_params(axis='y',labelsize=20)
plt.ylim(0, 10)  
plt.legend(title='income Category',fontsize=20)
plt.show()
```

<h1 style="font-family: 'Arial', sans-serif; font-size: 15px; direction: rtl; text-align: right; font-weight: bold;">
 החלטנו לבדוק יותר לעומק האם יש קשר בין המגדר+הגיל לרמת הרצינות 
</h1> 

```python
dating_df=dating_df.dropna(subset=['age']).copy()
bins = [18, 25, 35, 45, 56] 
labels = ['18-25', '25-35', '35-45', '45-55'] 
dating_df['age_group']=pd.cut(dating_df['age'],bins=bins, labels=labels, right=False)
filtereddata = dating_df.dropna(subset=['age_group', 'gender']) 
filtereddata['age_group'] = pd.cut(filtereddata['age'], bins=bins, labels=labels, right=False) 
grouped_goal_data = filtereddata.groupby(['gender', 'age_group'])['goal'].mean().unstack() 
grouped_count_data = filtereddata.groupby(['age_group', 'gender']).size().unstack().fillna(0) 
def absolute_value(val, gender_counts):
    a = int(val / 100. * sum(gender_counts)) 
    return a 
fig, axes = plt.subplots(3, 2, figsize=(20, 24))
plt.rcParams.update({'font.size':24})
grouped_goal_data.T.plot(kind='bar', ax=axes[0,0],color=['yellow','orange']) 
axes[0,0].set_title('Goals by Age and Gender', fontsize=24) 
axes[0,0].set_xlabel('Age Group',fontsize=24) 
axes[0,0].set_ylabel('Average Goals',fontsize=24) 
axes[0,0].legend(title='Gender', labels=['Female', 'Male']) 
axes[0,0].grid(True) 
axes[0,1].axis('off') 
for i, label in enumerate(labels):
    row=(i//2)+1
    col=i%2
    if label in grouped_count_data.index: 
        gender_counts = grouped_count_data.loc[label] 
        if gender_counts.sum() > 0: 
            gender_counts.plot(kind='pie', labels=['Female', 'Male'], autopct=lambda p: absolute_value(p , gender_counts), ax=axes[row,col], startangle=90,colors=['yellow','orange'],legend=False)
            axes[row,col].set_title(f'Participants Distribution Age {label}',fontsize=24) 
            axes[row,col].set_ylabel('') 

plt.tight_layout(pad=3.0,h_pad=5.0,w_pad=5.0) 
plt.show()
```
